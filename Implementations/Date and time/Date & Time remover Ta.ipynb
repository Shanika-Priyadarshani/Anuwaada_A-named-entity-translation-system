```python
import datetime
```

```python
def validateDate(date_text):
    try:
        (datetime.datetime.strptime(date_text, '%Y/%m/%d'))
        return 1
    except ValueError:
        try:
            (datetime.datetime.strptime(date_text, '%d/%m/%Y'))
            return 2
        except ValueError:
            try:
                (datetime.datetime.strptime(date_text, '%Y.%m.%d'))
                return 3
            except ValueError:
                try:
                    (datetime.datetime.strptime(date_text, '%d.%m.%Y'))
                    return 4
                except ValueError:
                    try:
                        (datetime.datetime.strptime(date_text, '%Y-%m-%d'))
                        return 5
                    except ValueError:
                        try:
                            (datetime.datetime.strptime(date_text, '%d-%m-%Y'))
                            return 6
                        except ValueError:
                            return False
```

```python
def validateTime(time_text):
    try:
        if(datetime.datetime.strptime(time_text, '%H:%M')):
            return 7
        elif(datetime.datetime.strptime(time_text, '%H:%M:%S')):
            return 8
    except ValueError:
        return False
```
